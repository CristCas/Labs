```python
#(Task 6.2.1)

"""
Переменная: Список
Функция: Сортировка списка от меньшего к большему
Сложность: O(N²)
"""

#O(N) + O(1) = O(N)
#O(N)
#O(1) + O(1) = 2 * O(1) = O(1)
#O(1) + O(1) = 2 * O(1) = O(1)
#O(1)

def foo(a):
  for i in range(len(a), 0, -1): #O(N) + O(1) = O(N)
      for j in range(1, i): #O(N)
          if a[j-1] > a[j]: #O(1) + O(1) = 2 * O(1) = O(1)
              a[j-1], a[j] = a[j], a[j-1] #O(1) + O(1) = 2 * O(1) = O(1)
  return a #O(1)

a = [1, 2, 3, 4, 2, 1, 3, 4, 3, 5, -2, -2, 2, -2, 65, 2, 2]
print(foo(a))
```

```python
#(Task 6.2.2)

"""
Переменная: Целое число
Функция: Int в Str (целое в строку)
Параметры: N = 1
Сложность: O(N)
"""

#O(1)
#2 * O(1)
#O(1)
#O(1)
#O(N) + O(1) = O(N)
#O(1) + O(1) + O(1) = O(1)
#O(1) + O(1) = O(1)
#O(1)
#O(1) + O(1) + O(N) * O(1) = 2 * O(1) + O(N) = O(1) + O(N) = O(N)

def foo(i):
  digits = "0123456789" #O(1)
  if i == 0: #2 * O(1)
      return "0"  #O(1)
  result = "" #O(1)
  while i > 0: #O(N) + O(1) = O(N)
      result = digits[i%10] + result #O(1) + O(1) + O(1) = O(1)
      i = i // 10#O(1) + O(1) = O(1)
  return result #O(1)

#O(1) + O(1) + O(N) * O(1) = 2 * O(1) + O(N) = O(1) + O(N) = O(N)
```

```python
#(Task 6.2.3)

"""
Переменная: Строка
Функция: Преобразовать цифры из строки в целое число
Параметры: N = len(s)
Сложность: O(N²)
"""

#O(1)
#O(N)
#O(N)
#O(1) + O(1) = O(1)
#O(1)
#O(1) + O(N) * O(N) * O(1) = O(N²)

def foo(s):
  val = 0 #O(1)
  for c in s: #O(N)
      if c.isdigit(): #O(N)
          val += int(c) #O(1) + O(1) = O(1)
  return val #O(1)

#O(1) + O(N) * O(N) * O(1) = O(N²)
```

```python
#(Task 6.2.4)

"""
Переменаая: Целое число
Функция: Вывести простые числа от 0 о указанного
Параметры: N = 1
Сложность: O(N²)
"""

#O(1)
#O(N) + O(1) = O(N)
#O(1)
#O(1)
#O(N) + O(1) + O(1) = O(N) + 2 * O(1) = O(N)
#O(1) + O(1) = 2 * O(1) = O(1)
#O(1)
#O(1)
#O(1) + O(1) = 2 * O(1) = O(1)
#O(1)
#O(1)
#O(1) + O(N) * (2 * O(1) + O(N) * 2 * O(1) * O(1)) = O(N²)

def foo(n):
  res = [] #O(1)
  for i in range(1, n + 1): #O(N) + O(1) = O(N)
      divisors = 0 #O(1)
      j = 2 #O(1)
      while j < i and divisors == 0: #O(N) + O(1) + O(1) = O(N) + 2 * O(1) = O(N)
          if i % j == 0: #O(1) + O(1) = 2 * O(1) = O(1)
              divisors += 1 #O(1)
          j += 1 #O(1)

      if divisors == 0: #O(1) + O(1) = 2 * O(1) = O(1)
          res.append(i) #O(1)
  return res #O(1)

  #O(1) + O(N) * (2 * O(1) + O(N) * 2 * O(1) * O(1)) = O(N²)
```

```python
#(Task 6.2.5)

"""
Переменная: Список
Функция: Проверка списка на чётные цифры
Параметры: 
N = len(nums)
Сложность: O(N)
"""

#O(N)
#O(1) + O(1) = 2 * O(1) = O(1)
#O(1)
#O(1)
#O(N) * O(1) * O(1) = O(N)

def foo(nums):
  for x in nums: #O(N)
      if x % 2 == 0: #O(1) + O(1) = 2 * O(1) = O(1)
          return True #O(1)
  else: #O(1)
      return False #O(N) * O(1) * O(1) = O(N)
```

```python
#(Task 6.2.7)

"""
Переменные: Целые числа
Функция: Угадать случайно заданное число
Параметры: N = 2
Сложность: O(N)
"""

#O(1)
#O(N)
#O(1) + O(1) = 2*O(1) = O(1)
#O(1)
#O(1)
#O(len(pointer) + O(1) = O(1) + O(1) = 2*O(1) = O(1)
#O(1)
#O(len(pointer) + O(1) = O(1) + O(1) = 2*O(1) = O(1)
#O(1)
#O(len(pointer) + O(1) = O(1) + O(1) = 2*O(1) = O(1)
#O(1)
#O(1)
#O(1)
#O(1) + O(N) * (O(1) + O(1) + O(1) + 3*O(1)*O(1)) = 
# = O(1) + O(N) * 6*O(1) = O(N)

def foo(low, high):
  guessing = True #O(1)
  while guessing: #O(N)
    guess = (low + high) // 2 #O(1) + O(1) = 2*O(1) = O(1)
    print("Загаданное число {0}?".format(guess)) #O(1)
    pointer = input( 
      "Введите '+', если Ваше число меньше.\n"
      "Введите '-' если Ваше число больше.\n"-
      "Введите '=', если я угадал.\n").lower() #O(1)
    if pointer == "+": #O(len(pointer) + O(1) = O(1) + O(1) = 2*O(1) = O(1)
      high = guess #O(1)
    elif pointer == "-": #O(len(pointer) + O(1) = O(1) + O(1) = 2*O(1) = O(1)
      low = guess #O(1)
    elif pointer == "=": #O(len(pointer) + O(1) = O(1) + O(1) = 2*O(1) = O(1)
      guessing = False #O(1)
    else: #O(1)
      print("Введите '+', '-' или '='.") #O(1)

  return guess

low, high = 0, 100
print("Пожалуйста, загадайте число от {0} до {1}!".format(low, high)) #O(len(pointer) + O(1) = O(1) + O(1) = 2*O(1) = O(1)
guess = foo(low, high) #O(1)
print("Игра окончена, Вы загадали число: {0}.".format(guess)) #O(1)

#O(1) + O(N) * (O(1) + O(1) + O(1) + 3*O(1)*O(1)) = 
# = O(1) + O(N) * 6*O(1) = O(N)
```
