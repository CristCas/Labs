```python
#(Task 6.2.2)

def foo(i):
    """!!!

    Параметры:
        - i (int): число.

    Сложность: !!!.
    """
    digits = "0123456789"
    if i == 0:
        return "0"
    result = ""
    while i > 0:
        result = digits[i%10] + result
        i = i // 10
    return result
```

```python
#(Task 6.2.3)

def foo(s):
    """!!!

    Параметры:
        - s (str): строка.

    Сложность: !!!.
    """
    val = 0
    for c in s:
        if c.isdigit():
            val += int(c)
    return val
```

```python
#(Task 6.2.4)

def foo(n):
    """!!!

    Параметры:
        - n (int): число.

    Сложность: !!!.
    """
    res = []
    for i in range(1, n + 1):
        divisors = 0
        j = 2
        while j < i and divisors == 0:
            if i % j == 0:
                divisors += 1
            j += 1

        if divisors == 0:
            res.append(i)

    return res
```
